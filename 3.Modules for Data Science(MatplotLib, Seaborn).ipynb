# MatplotLib 
- Data Visualisation & graphical plotting library for python & its numerical extension numpy
- Provides APIs for plottings in applications using GUI tools like Tkinter



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(10,20)
```

## Scatter Plot

```python
plt.scatter(x,y,c='r') #scatter
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("My first graph")
plt.savefig("my graph.png")
```

## Plot

```python
plt.plot(x,y,'b*-') #plot
```

## Subplot

```python
plt.subplot(1,3,1)
plt.plot(x,y,'go')


plt.subplot(1,3,2)
plt.plot(x,y,'b--')

plt.subplot(1,3,3)
plt.plot(x,y,'r*-')
```

## Sine curve

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y,c='r')
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.cos(x)
plt.plot(x,y,c='b')
```

```python
x = np.arange(0,4*np.pi,0.2)
y_s = np.sin(x)
y_c = np.cos(x)
              
plt.subplot(1,2,1)
plt.plot(x,y_s,'r')
              
plt.subplot(1,2,2)
plt.plot(x,y_c,'b')
```

## Bar Plot

```python
x1 = [4,5,6]
y1 = [9,5,8]

x2 = [9,10,12]
y2 = [10,20,30]

plt.bar(x1,y1,width=1,color='b')
plt.bar(x2,y2,color='g')
```

## Histogram

```python
a = [2,4,6,8,10,33,38,3,4,2,15,42,50,60,1,22,28] #To view the intensity
plt.hist(a)
```

## Boxplot

```python
data = [np.random.normal(0,i,100) for i in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data
```

## Pie Chart

```python
x = [200,130,90,30]
c = ['lightblue','yellowgreen','gray','pink']
l = ['python','C','C++','R']
e = [0.3,0,0,0]
plt.pie(x,explode=e,colors=c,labels=l,shadow=False)
```

# Seaborn

## Distribution Plots
- 1.Joint PLot
- 2.Pair Plot
- 3.Dist Plot

```python
import seaborn as sns
import pandas as pd
```

## Heatmap
- uses coloured cells to show correlation matrix between two discrete events

```python
df = pd.read_csv('/home/mahima/Downloads/datacor.csv')
```

```python
df.head(20)
```

```python
df.corr() #correlation
```

```python
df.dtypes
```

```python
sns.heatmap(df.corr())
```

## JoinPlot
- Shows the relationship between two numeric variables

```python
sns.jointplot(x='Duration',y='Calories',data=df,kind='hex',color='r')
```

```python
sns.jointplot(x='Duration',y='Calories',data=df,kind='reg')
```

## Pair Plot
- To show the relationship between more than two variables

```python
sns.pairplot(df,hue='Duration')
```

```python
sns.pairplot(df,hue='Duration')
```

## Dist Plot
- To check the distribution of the column feature

```python
sns.displot(df['Calories'],bins=10)
```
